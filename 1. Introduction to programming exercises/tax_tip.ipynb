```python
meal_cost = float(input("Enter cost of meal: "))

tax_rate = 0.05         
tip_rate = 0.18         

tax = meal_cost * tax_rate
tip = meal_cost * tip_rate
total = meal_cost + tax + tip

print("Tax amount: ₹%.2f" % tax)
print("Tip amount: ₹%.2f" % tip)
print("Total amount: ₹%.2f" % total)
```
